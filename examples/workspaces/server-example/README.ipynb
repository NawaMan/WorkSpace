# CodingBooth - Server Example

Simple HTTP server example demonstrating web service hosting in CodingBooth.

## Start Server
Run the following to start the server.

```bash
python -m http.server 8080 --bind 0.0.0.0 --directory ./www >/dev/null 2>&1 &
PID="$!"
```

# Check Server
You can simply visit : http://localhost:8080 to view the served page.
Or you can run the following.

```bash
curl -o /dev/null -s -w "%{http_code}\n" http://localhost:8080
```

## Stop  Server
Run the following to stop the server.

```bash
kill -9 $PID
```

Check again.

```bash
curl -o /dev/null -s -w "%{http_code}\n" http://localhost:8080
```
